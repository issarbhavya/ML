```python
from sklearn.datasets import make_circles
# Make 1000 examples
n_samples = 1000

# Create circles
X, y = make_circles(n_samples, 
                    noise=0.03, 
                    random_state=42)

# X : ndarray of shape (n_samples, 2)
# The generated samples.

# y : ndarray of shape (n_samples,)
# The integer labels (0 or 1) for class membership of each sample.
```

```python
import pandas as pd
circles = pd.DataFrame({"X0":X[:, 0], "X1":X[:, 1], "label":y})
circles
```

```python
# Visualize with a plot
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu)
```

```python
import tensorflow as tf
```

```python

tf.random.set_seed(42)
#create
model_1=tf.keras.Sequential([
    tf.keras.layers.Dense(1), ## takes input of one numpy array of size 2
    tf.keras.layers.Dense(10)
])

#compile
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(), ## binary since we are working with 2 clases (0 & 1)
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["accuracy"]) 

#fir
model_1.fit(X,y,epochs=5, verbose=0)
```

```python
model_1.evaluate(X, y)
```

```python
model_1.fit(X, y, epochs=200, verbose=0)
model_1.evaluate(X, y)
```

```python
def plot_decision_boundary(model, X, y):

   # Define the axis boundaries of the plot and create a meshgrid
   x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
   y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

    # The output of np.meshgrid() can be useful for creating grids of points for plotting
    # or for evaluating functions over a grid of points.
   xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))

   # what linspace do is : Return evenly spaced numbers over a specified interval.
   # np.linspace(2.0, 3.0, num=5)
   # array([2.  , 2.25, 2.5 , 2.75, 3.  ])
   
   ##### Create X values (we're going to predict on all of these)
   x_in = np.c_[xx.ravel(), yy.ravel()]


   # what np.c_ do is : 
   # np.c_[np.array([1,2,3]), np.array([4,5,6])]
   # array([[1, 4],
   #      [2, 5],
   #      [3, 6]])
  
   # what ravel do is :
   # x = np.array([[1, 2, 3], [4, 5, 6]])
   # np.ravel(x)
   # array([1, 2, 3, 4, 5, 6])

   ##### Make predictions using the trained model
   y_pred = model.predict(x_in)  

  # Check for multi-class
   if len(y_pred[0])> 1: 
     # checks the final dimension of the model's output shape, if this is > (greater than) 1, it's multi-class 
     print("doing multiclass classification...")
     # We have to reshape our predictions to get them ready for plotting
     y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)

     ## argmax : The function returns the indices of the maximum values along a specified axis in an array.

   else:
     print("doing binary classifcation...")
     y_pred = np.round(y_pred).reshape(xx.shape) 
     # first roundoff is done using round function
     # then, y_pred is brought to the same shape as of xx

   # Plot decision boundary
     plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7) # Contour: imag. line connecting points of eql val
     #contour and contourf draw contour lines and filled contours, respectively.

     plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu) # plots the 2 circles

    #  plt.xlim(xx.min(), xx.max()) # Get or set the x limits of the current axes.(here set)
    #  plt.ylim(yy.min(), yy.max())





## the below commented section is to understand working 

#    print("value of xx and yy is : ",xx) ## for yy everything will be similar
#    print("ravel values",xx.ravel())
# plot_decision_boundary(model_3,X,y)
```

```python
tf.random.set_seed(42)

# Create a model with a non-linear activation
model_5 = tf.keras.Sequential([
  tf.keras.layers.Dense(1, activation=tf.keras.activations.relu), # can also do activation='relu'
  tf.keras.layers.Dense(1) # output layer 
])

# Compile the model
model_5.compile(loss=tf.keras.losses.binary_crossentropy,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])

# Fit the model
history = model_5.fit(X, y, epochs=100)
```

```python
# Set random seed
tf.random.set_seed(42)

# Create a model
model_6 = tf.keras.Sequential([
  tf.keras.layers.Dense(4, activation=tf.keras.activations.relu), # hidden layer 1, 4 neurons, ReLU activation
  tf.keras.layers.Dense(4, activation=tf.keras.activations.relu), # hidden layer 2, 4 neurons, ReLU activation
  tf.keras.layers.Dense(1) # ouput layer
])

# Compile the model
model_6.compile(loss=tf.keras.losses.binary_crossentropy,
                optimizer=tf.keras.optimizers.Adam(lr=0.001), # Adam's default learning rate is 0.001
                metrics=['accuracy'])

# Fit the model
history = model_6.fit(X, y, epochs=100)
```

```python
model_6.evaluate(X, y)
```

```python
plot_decision_boundary(model_6, X, y)
```
